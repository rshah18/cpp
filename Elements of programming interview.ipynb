#### 1. write a program that takes an array A and an index i to A, and rearranges the elements such that all elements less than A[i] the pivot appear first, followed by elements equal to pivot, followed by elements greater than the pivot

* keep the following invariants during partitioning:
    * bottom group: A[0, smaller-1]; elements smaller than pivot
    * middle group: A[smaller, equal-1]; elements equal to pivot
    * unclassified group: A[equal, larger-1] all elements 
    * top group : A[larger, A.size()-1] elements greater than pivot
* initially all elements are unclassified 
* as we iterate we move elements into one of the categories 
* keep iterating as long as there are unclassified elements 
* unclassified elements exists as long as equal and larger dont cross each other 
* we start with one big unclassified array whose range is 0 to size-1, then as we find new elements we separate between 0 to smaller-1, smaller to equal-1, equal to larger-1. the bigger array vanishes and 3 other subarrays grow out of it.

* right answer

```cpp
void dutchflag(vector<int>& v, int p_index){
    int pivot = v[p_index]; 
    int smaller = 0; 
    int equal = 0; 
    int larger = v.size(); 
    
    //v[equal] is the incoming unclassified element

    while(equal < larger){
        if(v[equal] < pivot){
            swap(v[smaller++], v[equal++]);
        }
        else if(v[equal] == pivot){
            ++equal; //going more than smaller
        }
        else{
            //v[equal]>pivot
            swap(v[equal], v[--larger]); 
        }
    }
    print(v);
}
```
